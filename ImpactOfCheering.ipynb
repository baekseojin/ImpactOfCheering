```python
import pandas as pd
import matplotlib.pyplot as plt

# 폰트 설정
plt.rc('font', family='NanumBarunGothic')

# 엑셀 파일에서 데이터 불러오기
file_path = '/content/경기별 관중 데이터.xlsx'
df = pd.read_excel(file_path)

# 데이터 확인
df
```

```python
# 필요 없는 열 제거
df = df[['관중 수.1', '승점.1', '득점', '실점']]
df.columns = ['관중 수', '승점', '득점', '실점']

# 데이터 확인 (필요에 따라 주석 처리하거나 삭제 가능)
print(df.head())

# 산점도 그리기
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# 관중 수와 승점
ax[0].scatter(df['관중 수'], df['승점'], c='blue', label='Points')
ax[0].set_xlabel('관중 수')
ax[0].set_ylabel('승점')
ax[0].set_title('관중 수 - 승점')


# 관중 수와 득점
ax[1].scatter(df['관중 수'], df['득점'], c='green', label='Goals For')
ax[1].set_xlabel('관중 수')
ax[1].set_ylabel('득점')
ax[1].set_title('관중 수 - 득점')


# 관중 수와 실점
ax[2].scatter(df['관중 수'], df['실점'], c='red', label='Goals Against')
ax[2].set_xlabel('관중 수')
ax[2].set_ylabel('실점')
ax[2].set_title('관중 수 - 실점')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# 한글 폰트 설정
plt.rc('font', family='NanumBarunGothic')

# 엑셀 파일에서 데이터 불러오기
file_path = '/content/경기별 관중 데이터.xlsx'  # 파일 경로를 사용자가 제공한 엑셀 파일 경로로 변경하세요
df = pd.read_excel(file_path)

# 필요 없는 열 제거
df = df[['관중 수.1', '승점.1', '득점', '실점']]
df.columns = ['관중 수', '승점', '득점', '실점']

# 상관행렬 계산
corr_matrix = df.corr()

# 상관행렬 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('변수들 간의 상관행렬')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터 로드
data = {
    "팀명": ["부산", "수원삼성", "아산", "부천", "광주", "수원FC", "제주", "인천"],
    "관중 수": [4010, 10616, 3635, 2990, 5396, 4896, 5860, 10923],
    "승점": [0.50, 1.25, 1.38, 1.38, 1.13, 1.88, 1.50, 1.13],
    "득점": [1.25, 1.38, 1.50, 1.50, 1.75, 1.38, 1.13, 1.13],
    "실점": [2.13, 1.25, 1.13, 1.38, 1.88, 1.25, 1.25, 0.88]
}

df = pd.DataFrame(data)

# 산점도 그리기
fig, ax = plt.subplots(1, 3, figsize=(18, 5))

# 관중 수와 승점
ax[0].scatter(df['관중 수'], df['승점'], c='blue', label='승점')
ax[0].set_xlabel('관중 수')
ax[0].set_ylabel('승점')
ax[0].set_title('관중 수 - 승점')

# 관중 수와 득점
ax[1].scatter(df['관중 수'], df['득점'], c='green', label='득점')
ax[1].set_xlabel('관중 수')
ax[1].set_ylabel('득점')
ax[1].set_title('관중 수 - 득점')

# 관중 수와 실점
ax[2].scatter(df['관중 수'], df['실점'], c='red', label='실점')
ax[2].set_xlabel('관중 수')
ax[2].set_ylabel('실점')
ax[2].set_title('관중 수 - 실점')

plt.show()
```
